```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

def calculate_pixel_distance(video_path, frame_number=0, point1=(0, 0), point2=(0, 0)):
    # Open the video file
    cap = cv2.VideoCapture(video_path)

    # Set the frame to be read
    cap.set(cv2.CAP_PROP_POS_FRAMES, frame_number)

    # Read the frame
    ret, frame = cap.read()
    if not ret:
        print("Unable to read video frame")
        return

    # Display the frame with marked points
    cv2.circle(frame, point1, 5, (0, 255, 0), -1)
    cv2.circle(frame, point2, 5, (0, 255, 0), -1)
    cv2_imshow(frame)

    # Calculate the distance between the two points
    distance = np.sqrt((point2[0] - point1[0])**2 + (point2[1] - point1[1])**2)

    print(f"Pixel distance between the two points: {distance:.2f}")

    cap.release()

# Example usage with specified points
video_path = "/content/D27_Km015.0_T7820_D20231113_H073157_N6.mov"
calculate_pixel_distance(video_path, frame_number=100, point1=(50, 100), point2=(150, 200))
```
